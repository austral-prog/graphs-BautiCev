# Grafico 1: Numero de juegos completados
Este grafico muestra la cantidad de juegos completados al cabo de 100 dias. Elegi este grafico porque muestra claramente el aumento de los juegos completados al pasar el tiempo.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("Numero de juegos completados")
plt.xlabel("Tiempo en dias")
plt.ylabel("Juegos completados")
plt.grid(True)
plt.plot(np.random.random(100).cumsum())
plt.savefig('imgs/line_plot_time.png', dpi=300, bbox_inches='tight')
```

# Grafico 2: Diferencia de velocidades
Este grafico muestra la diferencia de tiempos de impresion entre dos impresoras 3d imprimiendo ambas el mismo objeto con los mismos parametros. Elegi el grafico de barras porque permite una mejor visualizacion de la diferencia de velocidades.

```python
impresoras = ['Ender 3S1','Bambu Lab p1s']
tiempo = [3.5,13.7]
bar_labels = ['Ender 3S1', 'Bambu Lab p1s']
bar_colors = ['tab:red', 'tab:green']

plt.bar(impresoras, tiempo, label=bar_labels, color=bar_colors)
plt.title("Tiempo de impresion")
plt.xlabel("Impresoras")
plt.ylabel("Tiempo en horas")
plt.grid(True)
plt.legend(title='Impresoras')
plt.savefig('imgs/bar_chart.png', dpi=300, bbox_inches='tight')
```

# Grafico 3: Costo de impresion 3d
Este grafico muestra el peso de las variables que influyen en el costo de una impresion 3d. Elegi este grafico porque muestra de forma obvia las influencias que tiene cada aspecto.

```python
variables = ["Material", "Electricidad", "Depreciacion de impresora", "Preparacion", "Post procesado"]
porcentaje = [50, 2, 16, 11, 22]
plt.pie(porcentaje, labels=variables, autopct='%1.1f%%')
plt.title("Costo de impresion")
plt.savefig('imgs/pie_chart.png', dpi=300, bbox_inches='tight')
```

# Grafico 4: Distribucion de alturas de capa en impresion 3D
Este grafico muestra la distribucion de las alturas de capas al imprimir en una impresora 3d. Elegi este grafico porque permite una observacion mas facil de los cambios y frecuencias de las alturas de capa.

```python
data = np.random.normal(loc=0.2, scale=0.002, size=1000)

plt.figure()
plt.hist(data, bins=30)

plt.grid(axis='y')
plt.title("Distribucion de alturas de capa en impresion 3D")
plt.xlabel("Altura de capa (mm)")
plt.ylabel("Frecuencia")

plt.savefig('imgs/histogram.png', dpi=300, bbox_inches='tight')
```

# Grafico 5: Horas de cursada por dia
Este grafico muestra las horas de cursada separadas en dias. Elegi este grafico porque muestra correctamente una comparacion entre los dias y la cantidad de horas.

```python
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]
horas = [2, 3, 8, 7, 7]

plt.figure(figsize=(10,5))
plt.plot(dias, horas, color='red', marker='o', linestyle="--", label='Horas')
plt.title("Horas de cursada")
plt.xlabel("Dias")
plt.ylabel("Horas")
plt.legend()
plt.grid(True)
plt.savefig('imgs/line_plot.png', dpi=300, bbox_inches='tight')
```
